## import libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pyLASCA.source import fileNames
from pyLASCA.source.lossBaseline import LossBaseline
```

## import dataframe

```python
file_name = "/Users/malkusch/PowerFolders/LaSCA/thr_alpha/SCTCAPS_thr_complete_noUnits_200821.csv"
df_data = pd.read_csv(file_name)
df_data.head()
```

## Threshold for allodynia mesurement after i.c. application of capsaicin

```python
df_data_temp = df_data[df_data["effect"] == "allodynia"]
df_data_temp = df_data_temp[df_data_temp["application_type"] == "i.c."]
df_data_temp = df_data_temp[df_data_temp["frame"] == 1]
df_data_temp = df_data_temp[df_data_temp["pxlSize_bl"] > 1.682e-01]
df_data_temp = df_data_temp[df_data_temp["pxlSize_bl"] <= 3.141e-01]
df_data_temp = df_data_temp[df_data_temp["pxlSize"] > 1.682e-01]
df_data_temp = df_data_temp[df_data_temp["pxlSize"] <= 3.141e-01]
df_data_temp = df_data_temp[df_data_temp["time"] > -15]
    
loss = LossBaseline(df_data_temp, x_min = 0.0, x_max = 13.0)
loss.minimize_brute()
loss.plot_minimization_process()
loss.thr_scatter_plot()
loss.plot_residues()
print("loss: %.3e at x: %.2f"%(loss.loss_value, loss.x_value))
```
