```python
from ipywidgets import HBox, VBox
import numpy as np
import cv2 as cv
from pyLASCA.gui.resolutionWidgets import ResolutionWidgets
from pyLASCA.source.fileNames import FileNames
from pyLASCA.source.frc import FRC
widgets = ResolutionWidgets()
fileNames = FileNames()
frc = FRC()
```

## Load Files
load two LaSCA FLUX images from the same measurement

```python
img1_box = VBox([widgets.pathText, widgets.pathButton])
img2_box = VBox([widgets.pathText2, widgets.pathButton2])
import_box = HBox([img1_box, img2_box])
display(import_box)
                  
```

## Analyze
perform FRC analysis on selected images

```python
analyze_box = VBox([widgets.pxlSizeTextFloat, widgets.analyzeButton])
display(analyze_box)

def runFrcAnalysis(event):
    widgets.clearOutput()
    display(analyze_box)
    frc.pxlSize = widgets.pxlSizeTextFloat.value
    frc.setImages(img1 = cv.imread(widgets.pathText.value, cv.IMREAD_GRAYSCALE),
                  img2 = cv.imread(widgets.pathText2.value, cv.IMREAD_GRAYSCALE))
    print("FLUX images")
    frc.plotImage()
    print("Power Spectra")
    frc.plotPowerSpectrum()
    print("FRC anlyis result")
    frc.plotFrcDist()
    print("Resolution:")
    print(frc)
    

widgets.analyzeButton.on_click(runFrcAnalysis)
```
